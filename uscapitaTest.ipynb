```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\aimva\OneDrive\Desktop\canada_per_capita_income.csv")
df
```

```python
%matplotlib inline
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
plt.plot(df.year, df['per capita income (US$)'], color='blue', marker='x')
plt.show()  # Display the plot
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
reg.predict([[2025]])
```

```python
d = pd.read_csv(r"C:\Users\aimva\OneDrive\Desktop\years(Sheet1).csv")
d
```

```python
# Ensure you're passing only the 'year' column for prediction
years_to_predict = d[['year']]  # This selects only the 'year' column

# Make predictions
p = reg.predict(years_to_predict)

# Add the predictions to your DataFrame
d['predicted income'] = p
d = d[['year', 'predicted income']]
d.to_csv('uspercapitaincome.csv', index=False)
d
```
